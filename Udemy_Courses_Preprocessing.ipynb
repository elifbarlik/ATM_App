```python
#Kurs Konularna Gre Dalm (Pasta Grafii):
konu_sayisi = df['konu'].value_counts() #konu sayisisini degiskene atar
plt.figure(figsize=(8, 8)) #8'e 8lik figure olusturur
plt.pie(konu_sayisi, labels=konu_sayisi.index, autopct='%1.1f%%', startangle=90) #konu_sayisi: veri, labels: etiketler, konu_sayisi.index=>kurs konulari,
#%1.1f%% : degerler bir ondalk basamakla ve yzde iaretiyle gorunur
plt.title('Kurs Konularna Gre Dalm') #grafik basligi
plt.show() #grafigi goster
```

```python
#cretli ve cretsiz Kurslarn Says (Bar Grafii):
plt.figure(figsize=(6, 4)) #6'ya 4'luk figure olsuturur
sns.countplot(x='ucretli_mi', data=df) #x='ucretli_mi'=> x ekseninde ucretli_mi sutunu kullanilacak, data: df =>veri seti df
plt.title('cretli ve cretsiz Kurslarn Says') #baslik
plt.xlabel('cretli') #x ekseni aciklama
plt.ylabel('Kurs Says')#y ekseni aciklama
plt.show() #grafigi goster
```

```python
#Kurs Seviyelerine Gre Abone Says (Kutu Grafii):
plt.figure(figsize=(8, 6)) #8'e 6'lik figure olsturur
sns.boxplot(x='seviye', y='abone_sayisi', data=df) #x ekseni icin seviye y ekseni icin abone_sayisi sutunu kullanilir, veri seti df
plt.title('Kurs Seviyelerine Gre Abone Says') #baslik
plt.xlabel('Kurs Seviyesi') #x ekseni etiket
plt.ylabel('Abone Says') #y ekseni etiket
plt.show() #goster
```

```python
#Kurs Sresi ve Fiyat Arasndaki liki (Scatter Plot):
plt.figure(figsize=(8, 6)) #8'e 6'lik figure
sns.scatterplot(x='toplam_sure', y='fiyat', data=df) #x ekseni icin toplam_sure, y ekseni icin fiyat sutunu kullanilir, veri seti df
plt.title('Kurs Sresi ve Kurs Fiyat Arasndaki liki') #baslik
plt.xlabel('Kurs Sresi') #x ekseni etiket
plt.ylabel('Kurs Fiyat') #y eksnei etiket
plt.show() #goster
```

```python
#Kurs Konularna Gre Ortalama Fiyat (Bar Grafii):
plt.figure(figsize=(10, 6)) #10'a 6'lik figure
sns.barplot(x='konu', y='fiyat', data=df, ci=None)  #x onu y fiyat sutunu, veri seti df,  ci=None gven araln kaldrr
plt.title('Kurs Konularna Gre Ortalama Fiyat') #baslik
plt.xlabel('Konu') #x ekseni etiket
plt.ylabel('Ortalama Fiyat') #y ekseni etiket
plt.xticks(rotation=45, ha='right')  # x eksenindeki etiketlerin daha iyi grnmesi iin x eksenindeki etiketler 45 derece dndrlr ve saa hizalanr
plt.tight_layout()  # grafiin dzgn bir ekilde yerlemesi iin
plt.show()# goster
```
